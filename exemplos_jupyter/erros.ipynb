# __Erro Absoluto:__

---

## Exemplos 1: Casos escalares

```python
# Calcula o erro absoluto para cada par de valores abaixo
valores = [(10.0 , 7,0),
           (3.5 , 3.5),
           (3.5 , -3.5),
           (-4.0 , -6.0)]

for par in valores:
    v1,v2 = par[0], par[1]
    ea = erro_absoluto(v1,v2)
    print(f'Erro absoluto entre ({v1} , {v2}) = {ea}')
```

---

## Exemplos 2: Arrays NumPy

```python
# Calcula o erro absoluto para cada par de arrays NumPy abaixo
arrays = [( np.array([100,20,5]), np.array([99,21,5]) ),
          ( np.array([10,-5.0]) , np.array([8,-4.5]) ),
          ( np.array([1.1,2.2,3.3]) , np.array([1.0,2.0,4.0]) )]

for par in arrays:
    arr1, arr2 = par[0], par[1]
    ea = erro_absoluto(arr1, arr2)
    print(f'Erro absoluto entre ({arr1} , {arr2}) = {ea}')
```

# __Erro relativo:__

---

## Exemplos 1: Casos escalares

```python
valores = [(100.0 , 99,0),
           (50.0 , 60.0),
           (-10.0 , -11.0),
           (-4.0 , -6.0)]

for par in valores:
    v1, v2 = par[0], par[1]
    er = erro_relativo(v1, v2)
    print(f'Erro relativo entre ({v1} , {v2}) = {er}')
```

```python
vetores_de_teste = [
    (np.array([100.0, 20.0, 5.0]), np.array([99.0, 21.5, 5.0])),
    (np.array([10.0, -5.0]), np.array([8.0, -4.5])),
    (np.array([1.1, 2.2, 3.3]), np.array([1.0, 2.0, 4.0])),
    (np.array([100.0, 0.0, 0.0, -10.0]), np.array([99.0, 0.0, 10.0, -11.0]))
]

for vr, va in vetores_de_teste:
    er = erro_relativo(vr, va)
    print(f"ER: {er} \t| VR: {vr} \t| VA: {va}")
```
